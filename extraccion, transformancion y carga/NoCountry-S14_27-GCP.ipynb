Query fights cargo

```python
query = """
  SELECT
    f.Anio,
    f.Mes,
    f.AeropuertoO,
    a1.Ciudad as CiudadO,
    s.Estado as EstadoO,
    f.AeropuertoD,
    a2.Ciudad as CiudadD,
    a2.Pais as PaisD,
    f.Aerolinea,
    a.NombreAerolinea,
    a.TipoAerolinea,
    SUM(f.VCargaC) as VCargaC,
    SUM(f.VCargaP) as VCargaP
  FROM
    cloudtest-400719.NoCountry_s14_27.flightsCargo f
  LEFT OUTER JOIN
    cloudtest-400719.NoCountry_s14_27.airports a1 ON f.IDAeropuertoO = a1.IDAeropuerto
  LEFT OUTER JOIN
    cloudtest-400719.NoCountry_s14_27.airports a2 ON f.IDAeropuertoD = a2.IDAeropuerto
  LEFT OUTER JOIN
    cloudtest-400719.NoCountry_s14_27.airlines a ON f.IDAerolinea = a.IDAerolinea
  LEFT OUTER JOIN
    cloudtest-400719.NoCountry_s14_27.states s ON a1.Pais = s.IDEstado
  GROUP BY
    f.Anio,
    f.Mes,
    f.AeropuertoO,
    a1.Ciudad,
    s.Estado,
    f.AeropuertoD,
    a2.Ciudad,
    a2.Pais,
    f.Aerolinea,
    a.NombreAerolinea,
    a.TipoAerolinea
  HAVING anio > 2018
"""
```

```python
job = client.query(query)
dfFC = job.to_dataframe()
```

```python
query = """
  SELECT
    f.Anio,
    f.Mes,
    f.AeropuertoO as AeropuertoOrigen,
    a1.NombreAeropuerto as AeropuertoOrigenNombre,
    a1.Ciudad as CiudadOrigen,
    s.Estado as EstadoOrigen,
    f.AeropuertoD as AeropuertoDestino,
    a2.NombreAeropuerto as AeropuertoDestinoNombre,
    a2.Ciudad as CiudadDestino,
    a2.Pais as PaisDestino,
    f.Aerolinea,
    a.NombreAerolinea as AerolineaNombre,
    SUM(f.VPasajerosC) as VPasajerosC,
    SUM(f.VPasajerosP) as VPasajerosP,
    SUM(f.AsientosC) as AsientosC,
    SUM(f.AsientosP) as AsientosP,
    SUM(f.PasajerosC) as PasajerosC,
    SUM(f.PasajerosP) as PasajerosP
  FROM
    cloudtest-400719.NoCountry_s14_27.flights f
  LEFT OUTER JOIN
    cloudtest-400719.NoCountry_s14_27.airports a1 ON f.IDAeropuertoO = a1.IDAeropuerto
  LEFT OUTER JOIN
    cloudtest-400719.NoCountry_s14_27.airports a2 ON f.IDAeropuertoD = a2.IDAeropuerto
  LEFT OUTER JOIN
    cloudtest-400719.NoCountry_s14_27.airlines a ON f.IDAerolinea = a.IDAerolinea
  LEFT OUTER JOIN
    cloudtest-400719.NoCountry_s14_27.states s ON a1.Pais = s.IDEstado
  GROUP BY
    f.Anio,
    f.Mes,
    f.AeropuertoO,
    a1.NombreAeropuerto,
    a1.Ciudad,
    s.Estado,
    f.AeropuertoD,
    a2.NombreAeropuerto,
    a2.Ciudad,
    a2.Pais,
    f.Aerolinea,
    a.NombreAerolinea
  HAVING f.AeropuertoO = 'MIA' and f.Aerolinea = 'AA'
"""
```
